## Barplot 
the barplot represents categorial data on one axis and continous data on other axis

syntax
plt.bar(x,y,color,label,edgecolor,)

```python
df=pd.read_csv('spotify-2023.csv', encoding='latin')
df
```

```python
df1=df.groupby('released_year')['in_spotify_playlists'].mean().reset_index()
df1=df1.sort_values(by='released_year',ascending=False).head(10)
```

```python
plt.Figure(figsize=(12,6))
plt.bar(df1['released_year'],df1['in_spotify_playlists'],color='lightgreen',edgecolor='black')
plt.xlabel('Released Year')
plt.ylabel('in_spotify_playlists Songs')
plt.show()
```

```python
df1=df.groupby('released_year')['in_apple_playlists','in_spotify_charts','in_spotify_playlists'].mean().reset_index()
df1=df1.sort_values(by='released_year',ascending=False).head(10)
```

```python
plt.Figure(figsize=(12,6))
plt.bar(df1['released_year'],df1['in_apple_playlists'],color='lightgreen',label='in_apple_playlist')
plt.bar(df1['released_year'],df1['in_spotify_charts'],color='red',label='in_spotify_charts',alpha=0.4)
plt.legend()
plt.xlabel('released_Year')
plt.show()
```

```python
df2=df1.head(10)
df2=df2.set_index('released_year')
```

```python

fig = plt.figure(figsize=(12,6)) # setting plt figure
ax = fig.add_subplot(111) 
ax2=ax.twinx()
df2['in_apple_playlists'].plot(kind='bar',color='red',ax=ax,width=0.4,position=0)
df2['in_spotify_charts'].plot(kind='bar',color='green',ax=ax2,width=0.4,position=1)
ax.set_ylabel('in_apple_playlists')
ax2.set_ylabel('in_spotify_charts')
```

## Histogram  
Used for frequency distribution of given data

```python
#Create a simple Histogram for â€˜Total_marksâ€™
plt.hist(marks['Total_marks'],edgecolor='black')
plt.show()
```

```python
#Create a Histogram for 'Total_marks' and apply the Following customization to the plot.
# â—Make the Figuresize(9,4)
# â—Make Histogram color'teal' 
# â—Make Edgecolor as'black'
# â—Definebins for Histogram as bins=10 
# â—Add a title to the Plot with fontsize=15 and color=purple
# â—AddX-Labelas'TotalMarks'
# â—Also Show theplot
plt.figure(figsize=(9,4))
plt.hist(marks['Total_marks'],color='teal',bins=10,edgecolor='black')
plt.title("Frequency Distribution of Total Marks", fontsize=15,color='purple')
plt.xlabel("Total Marks")
plt.show()
```

## Pie Chart 
Used for Percentage distribution 

```python
# Create a Piechart of 'Namewise Python Marks' and apply the Following customization to thep lot.
# â—Make the Figuresize(5,5)
# â—Define pie plot with x as'Python'marks and lableas'Name'
# â—Pass the color list for colour of pie chart sections as colors=['red','cornflowerblue','yellow','tan','brown','green','cyan','pink']
# â—Define autopact till two decimal points as autopct='%.2f%%'
# â—Make shadow=True
# â—Make Counterclockwise=False
# â—Define color of the label using txt props as'navy'color.
# â—Add a title to the Plot with font size=15andcolor='teal'
# â—AlsoShowtheplot

plt.figure(figsize=(5,5))
plt.pie(marks['Python'],labels=marks['Name'],colors=['red','cornflowerblue','yellow','tan','brown','green','cyan','pink'],
        autopct='%.2f%%',shadow=True,counterclock=False,textprops=dict(color="navy"),explode=(0,0,0,0.1,0,0,0,0))
plt.title("Percentage Distribution of Pyhton Namewise")
```

```python
#Create a Pie chart of 'Namewise TotalMarks'and apply the Following customization to the plot

plt.figure(figsize=(5,5))
plt.pie(marks['Total_marks'],labels=marks['Name'], colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],
        autopct='%.2f%%',explode=(0,0,0,0,0.1,0,0,0.2),counterclock=False,shadow=True)
plt.legend(marks['Name'],title="Names",loc='right',bbox_to_anchor=(1,1,0.6,-0.5))
plt.title("Percentage Distribution of Total Marks w.r.t Name",fontsize=15,color='blue')
plt.show()
```

## Boxplot 
used to represent quartile / percentile distribution.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Bollywood_Movies_Dataset.csv')
df['BO_Collection'] = df['BO_Collection'].astype(float)
df['BO_Collection'] = df['BO_Collection'].astype(int) 
df['Budget'] = df['Budget'].astype(int)
# Making a new column 'Profit' by subtracting 'Budget'from 'BO_Collection'
df['Profit'] = df['BO_Collection']-df['Budget']
df.head()
```

```python
df1=pd.read_csv('DataScience_Marks_Dataset.csv')
df1
```

```python
plt.boxplot(df['BO_Collection'])
plt.grid()
```

From the Plot we can conclude  that <br>
- q1 - 8.00  i.e 25 %
- q2 - 23.50  i.e 50 % median
- q3 - 56.00  i.e 75 %
- max - 120 the values less than max value
- min - 1 

```python
plt.boxplot(df1['Total_marks'])
plt.grid()
plt.show()
```

```python
plt.boxplot(df['Budget'])
plt.grid()
plt.show()
```

```python
fig=plt.figure(1,figsize=(9,6))
ax=fig.add_subplot(111)
ax.boxplot(df['Budget'],vert=False,manage_ticks=True)
ax.set_xlabel('Value')
ax.set_yticks([1])
ax.set_yticklabels(['Budget'])
quantl=np.quantile(df['Budget'],np.array([0.00,0.25,0.50,0.75,1.00]))
ax.vlines(quantl,[0]*quantl.size,[1]*quantl.size,color='b',ls=':',lw=0.5,zorder=0)
ax.set_ylim(0.5,1.5)
ax.set_xticks(quantl)
plt.show()
```
